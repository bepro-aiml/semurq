We'll plot the distribution of `Global_Sales` (millions of copies sold worldwide).

### 📚 Syntax

```python

```

- `values` — a list / Series of numbers (one per row).

## 5. 🪵 The Log-Scale Moment

**From the lecture:** *plotting income linearly with a few billionaires flattens everyone else. `plt.yscale('log')` reveals the true distribution shape.*

Same problem with our data: a few mega-selling games crush the chart. **Solution:** make the y-axis logarithmic. On a log scale, each tick is 10× the previous one (1, 10, 100, 1000, ...) — small values become visible.

### 📚 Syntax

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
```

- `plt.subplots(rows, cols)` — make a grid of charts. Here: 1 row × 2 columns.
- It returns **two** things: `fig` (the whole figure) and `axes` (a list of the individual chart areas).
- `axes[0]` is the FIRST (left) chart. `axes[1]` is the SECOND (right) chart.
- `figsize=(12, 4)` means 12 inches wide, 4 inches tall.

````python
# ────────────────────────────────────────────────────────────
# Create a figure with two charts side-by-side.
# ────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# ─── LEFT chart: linear y-axis (the default) ──────────────────
axes[0].hist(df['Global_Sales'], bins=50)   # histogram on chart #0
axes[0].set_title('Linear y-axis')          # title for chart #0
axes[0].set_xlabel('Global Sales (millions)')
axes[0].set_ylabel('Game count')

# ─── RIGHT chart: log y-axis (the magic) ──────────────────────
axes[1].hist(df['Global_Sales'], bins=50)   # same histogram on chart #1
axes[1].set_yscale('log')                   # ← change the y axis to LOG
axes[1].set_title('Log y-axis')
axes[1].set_xlabel('Global Sales (millions)')
axes[1].set_ylabel('Game count (log)')

# tight_layout() prevents labels overlapping each other.
plt.tight_layout()
plt.show()
````

## 6. 📦 Chart 3 — Boxplot: Sales by Genre

**Concept.** A **boxplot** shows median, quartiles, and outliers for each group. Perfect for comparing one numeric variable across categories.

Anatomy of a box:
- The **box** is the middle 50% of the values (Q1 to Q3).
- The **line inside the box** is the **median** (middle value).
- The **whiskers** extend to the typical range.
- The **dots above** are **outliers** — unusually high values.

Question: *which game genre has the highest median global sales?*

````python
# ────────────────────────────────────────────────────────────
# First, filter out the tiny obscure games. We only keep games
# that sold at least 0.1 million (= 100,000) copies. Otherwise
# the chart is full of junk and the boxes are squashed flat.
#
# This is BOOLEAN FILTERING from Class 4:
#   df[ df['column'] >= value ]
# ────────────────────────────────────────────────────────────
popular = df[df['Global_Sales'] >= 0.1].copy()
# .copy() makes 'popular' an independent DataFrame. Without it,
# Python sometimes shows a SettingWithCopyWarning. .copy() is safe.

# ────────────────────────────────────────────────────────────
# Build the figure manually (OO-style — see section 9).
# figsize=(10, 5) means 10 inches wide, 5 inches tall.
# ────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(10, 5))

# ────────────────────────────────────────────────────────────
# sns.boxplot draws one box per Genre (the X axis is categorical).
# y='Global_Sales' is the numeric column we summarise.
# ax=ax tells Seaborn to draw on OUR figure, not a new one.
# ────────────────────────────────────────────────────────────
sns.boxplot(popular, x='Genre', y='Global_Sales', ax=ax)

# Use a log Y axis so the mega-hits don't squash the boxes.
ax.set_yscale('log')

# Title and Y label.
ax.set_title('Global Sales by Genre (log scale)')
ax.set_ylabel('Global Sales (millions, log)')

# Genre names are long — rotate them so they don't overlap.
# rotation=30 = tilt by 30 degrees.
# ha='right' = align text to the RIGHT side of the tick.
plt.xticks(rotation=30, ha='right')

plt.tight_layout()
plt.show()
````

## 7. ⚡ Chart 4 — Scatter Plot: NA vs EU Sales, Coloured by Genre

**Concept.** A **scatter plot** shows the relationship between two numeric variables — each dot is one row of data. Adding `hue=` colours each dot by a third (categorical) variable. So you can see THREE dimensions in one chart.

### 📚 Syntax

```python
sns.scatterplot(df, x='col_a', y='col_b', hue='category', alpha=0.6)
```

- `x` and `y` — the two numeric columns.
- `hue` — the categorical column for colour.
- `alpha` — transparency from 0 (invisible) to 1 (solid). Useful when many dots overlap.

````python
# ────────────────────────────────────────────────────────────
# Build the canvas. figsize=(9, 6) = 9 wide × 6 tall.
# ────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(9, 6))

# ────────────────────────────────────────────────────────────
# sns.scatterplot draws one dot per row.
#   - x = NA_Sales         (horizontal axis)
#   - y = EU_Sales         (vertical axis)
#   - hue = Genre          (each genre gets its own colour)
#   - alpha = 0.6          (60% opaque so overlaps are visible)
#   - ax = ax              (draw on our figure)
# We use 'popular' (the filtered DataFrame from above) so the
# chart isn't cluttered with thousands of near-zero dots.
# ────────────────────────────────────────────────────────────
sns.scatterplot(
    popular,
    x='NA_Sales',
    y='EU_Sales',
    hue='Genre',
    alpha=0.6,
    ax=ax
)

ax.set_title('North America vs Europe Sales (millions)')
ax.set_xlabel('North America Sales (M)')
ax.set_ylabel('Europe Sales (M)')
plt.tight_layout()
plt.show()
````

## 8. 🌡️ Chart 5 — Correlation Heatmap

**Concept.** A **heatmap** shows a 2D grid of values as colours. For correlations, the grid is square (each numeric variable vs each other numeric variable) and the colour shows how strongly they move together.

- 🔴 **Red / +1** — perfectly positive correlation (both rise together)
- ⚪ **White / 0** — no correlation
- 🔵 **Blue / -1** — perfectly negative correlation (one rises, the other falls)

### 📚 Syntax

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

- `df.corr(numeric_only=True)` — compute the correlation between every pair of numeric columns.
- `annot=True` — write the correlation number inside each square.
- `cmap='coolwarm'` — the colour palette (blue ↔ red).

````python
# ────────────────────────────────────────────────────────────
# Build the figure first.
# ────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 6))

# ────────────────────────────────────────────────────────────
# Step 1: compute the correlation matrix.
# numeric_only=True tells pandas to skip text columns (Name, Genre, ...).
# Correlation only makes sense for numbers.
# ────────────────────────────────────────────────────────────
corr_matrix = df.corr(numeric_only=True)

# ────────────────────────────────────────────────────────────
# Step 2: draw the heatmap.
#   annot=True   → write each number inside its square
#   fmt='.2f'    → format numbers with 2 decimal places (e.g. 0.97)
#   cmap='coolwarm' → red↔blue palette
#   vmin=-1, vmax=1 → lock colour scale (so 0 is always white)
#   ax=ax        → draw on our figure
# ────────────────────────────────────────────────────────────
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    vmin=-1, vmax=1,
    ax=ax
)
ax.set_title('Correlation between numeric columns')
plt.tight_layout()
plt.show()
````

## 9. 🎨 Chart 6 — OO Matplotlib (your storytelling chart)

**Concept.** The **OO (Object-Oriented) Matplotlib** pattern gives you fine-grained control. You create the figure (`fig`) and the chart area (`ax`) yourself, then customise everything by calling methods on `ax`.

### 📚 The pattern

```python
fig, ax = plt.subplots(figsize=(8, 5))     # create canvas + chart
ax.plot(x, y)                              # draw something
ax.set_xlabel('x')                         # label the X axis
ax.set_title('Squares')                    # add a title
fig.savefig('out.png', dpi=150)            # save to file
```

This chart is your **explanatory chart for the day** — the one you'd show to a non-technical manager. It must follow the *Five Rules for Honest Charts*.

````python
# ────────────────────────────────────────────────────────────
# We want the Top 10 best-selling games of all time.
#
# .sort_values('Global_Sales', ascending=False) sorts the whole
# DataFrame by Global_Sales, BIGGEST first.
# .head(10) takes the first 10 rows of the sorted result.
# ────────────────────────────────────────────────────────────
top10 = df.sort_values('Global_Sales', ascending=False).head(10)

# Create the canvas (10 wide × 6 tall).
fig, ax = plt.subplots(figsize=(10, 6))

# ────────────────────────────────────────────────────────────
# barh = HORIZONTAL bar chart (good when category names are long).
#   - First argument: labels for the bars (game names)
#   - Second argument: lengths of the bars (sales numbers)
#   - color: hex colour code for the bars
# ────────────────────────────────────────────────────────────
ax.barh(top10['Name'], top10['Global_Sales'], color='#3498db')

# By default barh puts the FIRST row at the bottom — flip it so
# the biggest bar is on top, like a normal leaderboard.
ax.invert_yaxis()

ax.set_xlabel('Global Sales (millions of copies)')
ax.set_title('🏆 Top 10 Best-Selling Games of All Time')

# ────────────────────────────────────────────────────────────
# Add a number label at the end of each bar.
# enumerate(...) gives us (index, value) for each item.
# ────────────────────────────────────────────────────────────
for i, v in enumerate(top10['Global_Sales']):
    # ax.text(x, y, text, va='center')
    #   x = position along the bar (just past the end)
    #   y = the bar's row index
    #   text = the number formatted with 1 decimal + 'M'
    #   va='center' = vertically centre the text on the bar
    ax.text(v + 0.5, i, f'{v:.1f}M', va='center')

plt.tight_layout()

# Save the chart to disk as a PNG image. dpi=150 = decent quality.
fig.savefig('top10_games.png', dpi=150)

plt.show()
````

### 🎓 Vocab Recap

- **Histogram** — distribution of one numeric column.
- **KDE** — smooth density curve over a histogram.
- **Boxplot** — median, quartiles, outliers per group.
- **Scatter plot** — relationship between two numeric columns.
- **Heatmap** — colour-coded 2D grid.
- **Hue** — third variable shown by colour.
- **Alpha** — transparency (0 = invisible, 1 = solid).
- **Log scale** — compresses big values, reveals small ones.
- **`fig, ax = plt.subplots()`** — the OO pattern: *fig* is the canvas, *ax* is the chart.
- **Anscombe's quartet** — four datasets with identical stats, very different shapes. *Always plot before trusting summaries.*
