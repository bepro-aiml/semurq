## 3. 📊 Chart 1 — Histogram with `plt.hist` (plain Matplotlib)

**Concept.** A **histogram** buckets continuous values into bins and counts how many fall in each bin. It's the FIRST chart to draw for any numeric column — it shows the *shape* of the distribution.

We'll plot the distribution of `Global_Sales` (in millions of copies).

````python
# ────────────────────────────────────────────────────────────
# plt.hist(values, bins=N) draws a histogram with N bins.
# More bins = finer resolution but noisier.
# ────────────────────────────────────────────────────────────
plt.hist(df['Global_Sales'], bins=50)
plt.title('Global Sales distribution (Matplotlib)')
plt.xlabel('Global Sales (millions of copies)')
plt.ylabel('Number of games')
plt.show()
````

## 4. 📊 Chart 2 — Histogram with `sns.histplot` (Seaborn) + KDE

**Concept.** Seaborn's version is shorter, prettier, and you can overlay a **KDE (Kernel Density Estimate)** — a smooth curve that approximates the distribution shape.

````python
# ────────────────────────────────────────────────────────────
# Same chart in Seaborn. kde=True adds the smooth density line.
# Notice: Seaborn understands DataFrames natively — pass df + column name.
# ────────────────────────────────────────────────────────────
sns.histplot(df, x='Global_Sales', kde=True, bins=50)
plt.title('Global Sales distribution (Seaborn + KDE)')
plt.show()
````

## 5. 🪵 The Log-Scale Moment

**From the lecture:** *plotting income linearly with a few billionaires flattens everyone else. `plt.yscale('log')` reveals the true distribution shape.*

Same problem here: a few mega-selling games crush the chart. Let's compare linear vs log y-axis side by side.

````python
# ────────────────────────────────────────────────────────────
# plt.subplots(1, 2, ...) creates 1 row × 2 columns of charts.
# axes[0] is the left chart; axes[1] is the right chart.
# ────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(df['Global_Sales'], bins=50)
axes[0].set_title('Linear y-axis')
axes[0].set_xlabel('Global Sales (millions)')
axes[0].set_ylabel('Game count')

axes[1].hist(df['Global_Sales'], bins=50)
axes[1].set_yscale('log')   # ← THE log magic
axes[1].set_title('Log y-axis')
axes[1].set_xlabel('Global Sales (millions)')
axes[1].set_ylabel('Game count (log)')

plt.tight_layout()
plt.show()
````

## 6. 📦 Chart 3 — Boxplot: Sales by Genre

**Concept.** A **boxplot** shows median, quartiles, and outliers for each group. Perfect for comparing a numeric distribution across categories.

Question: *which game genre has the highest median global sales?*

````python
# ────────────────────────────────────────────────────────────
# We focus on games that sold at least 0.1M copies and use a log y-axis,
# otherwise the mega-hits squash the boxes flat.
# ────────────────────────────────────────────────────────────
popular = df[df['Global_Sales'] >= 0.1].copy()   # boolean filter from Class 4 ✨

fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(popular, x='Genre', y='Global_Sales', ax=ax)
ax.set_yscale('log')
ax.set_title('Global Sales by Genre (log scale)')
ax.set_ylabel('Global Sales (millions, log)')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
````

## 7. ⚡ Chart 4 — Scatter Plot: NA vs EU Sales, Coloured by Genre

**Concept.** A **scatter plot** shows the relationship between two numeric variables. Adding `hue=` colours each dot by a third (categorical) variable — three dimensions in one chart.

````python
fig, ax = plt.subplots(figsize=(9, 6))
sns.scatterplot(
    popular,
    x='NA_Sales',
    y='EU_Sales',
    hue='Genre',
    alpha=0.6,           # alpha = transparency, helps when dots overlap
    ax=ax
)
ax.set_title('North America vs Europe Sales (millions)')
ax.set_xlabel('North America Sales (M)')
ax.set_ylabel('Europe Sales (M)')
plt.tight_layout()
plt.show()
````

## 8. 🌡️ Chart 5 — Correlation Heatmap

**Concept.** A **heatmap** shows a 2D grid of values as colours. For correlations, the grid is square (each variable vs each other variable) and the colour shows how strongly they move together.

- 🔴 **Red / +1** — perfectly positive correlation (both rise together)
- ⚪ **White / 0** — no correlation
- 🔵 **Blue / -1** — perfectly negative correlation (one rises, other falls)

````python
# ────────────────────────────────────────────────────────────
# .corr(numeric_only=True) gives the correlation matrix between
# every pair of numeric columns. annot=True writes the number
# inside each square. cmap='coolwarm' is the red↔blue palette.
# ────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    fmt='.2f',           # 2 decimal places
    cmap='coolwarm',
    vmin=-1, vmax=1,     # lock the colour scale to [-1, 1]
    ax=ax
)
ax.set_title('Correlation between numeric columns')
plt.tight_layout()
plt.show()
````

## 9. 🎨 Chart 6 — OO Matplotlib (your storytelling chart)

**Concept.** The **OO (Object-Oriented) Matplotlib** pattern gives you fine-grained control: you create the figure and axes manually, then customise everything.

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(x, y)
ax.set_xlabel('x'); ax.set_title('Squares')
fig.savefig('out.png', dpi=150)
```

This chart is your **explanatory chart for the day** — the one you would show to a non-technical manager. It must follow the *Five Rules for Honest Charts*.

````python
# ────────────────────────────────────────────────────────────
# Top-10 best-selling games of all time. We sort by Global_Sales
# and take the head, then plot a horizontal bar chart for readability.
# ────────────────────────────────────────────────────────────
top10 = df.sort_values('Global_Sales', ascending=False).head(10)

fig, ax = plt.subplots(figsize=(10, 6))
ax.barh(top10['Name'], top10['Global_Sales'], color='#3498db')
ax.invert_yaxis()                                # biggest bar on top
ax.set_xlabel('Global Sales (millions of copies)')
ax.set_title('🏆 Top 10 Best-Selling Games of All Time')
for i, v in enumerate(top10['Global_Sales']):    # label each bar
    ax.text(v + 0.5, i, f'{v:.1f}M', va='center')
plt.tight_layout()
fig.savefig('top10_games.png', dpi=150)          # save as image
plt.show()
````

### 🎓 Vocab Recap

- **Histogram** — distribution of one numeric column.
- **KDE** — smooth curve estimate of a distribution.
- **Boxplot** — median, quartiles, outliers per group.
- **Scatter plot** — relationship between two numeric columns.
- **Heatmap** — colour-coded 2D grid.
- **Hue** — third variable shown by colour.
- **Alpha** — transparency (0 = invisible, 1 = solid).
- **Log scale** — compresses big values, reveals small ones.
- **Anscombe's quartet** — four datasets with identical stats, very different shapes. *Always plot before trusting summaries.*
