```python
def calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    return total / count

test1 = [10 ,20, 30, 40, 50]
test2 = [15, 25, 35, 45, 55]
print('Average 1: ', calculate_average(test1))
print('Average 2: ', calculate_average(test2))
```

```python
import random
exam_scores = [random.randint(0, 100) for student_scores in range(10)]
print("Scores: ", exam_scores)

avg = calculate_average(exam_scores)
print(f"Class average: {avg:.1f}")

if avg >= 60:
    print("Result: Passed with minimum score")
else:
    print("Result: Failed")
```
