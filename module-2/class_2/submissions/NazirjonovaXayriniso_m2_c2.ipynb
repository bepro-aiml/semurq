```
import pandas as pd # [cite: 11, 19]

# Updated loading line with encoding fix
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1') # [cite: 20, 22]

# Display the first 10 rows [cite: 20]
df.head(10)
```

I imported the Pandas library and loaded the dataset using pd.read_csv. I applied 'latin1' encoding to ensure the file read correctly, and used .head(10) to verify that columns like Ship Mode and Segment are visible

I examined the dataset and found it contains 9,994 rows and 21 columns. The statistics show a wide range of Sales and Profit values, and I used .isnull().sum() to identify which columns require cleaning.

```
# Filling numerical columns with median [cite: 43, 45, 53]
df['Sales'].fillna(df['Sales'].median(), inplace=True)

# Filling categorical columns with mode [cite: 46, 48, 53]
df['Ship Mode'].fillna(df['Ship Mode'].mode()[0], inplace=True)

# Verification [cite: 49]
print(df.isnull().sum())
```

I handled missing values by filling numerical columns with the median, which is more robust against outliers, and categorical columns with the mode. I then verified that the null count for all columns is now zero.

```
print("Duplicate rows found:", df.duplicated().sum()) # Count them
df = df.drop_duplicates() # Remove them
print("New Shape:", df.shape)
```

I checked for redundant data using .duplicated(). After finding the duplicate rows, I removed them using .drop_duplicates() to ensure our analysis isn't skewed by repeated entries.

```
# Convert text dates to datetime objects
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Verify the change
print(df.dtypes)
```

I converted 'Order Date' and 'Ship Date' from strings to datetime objects using pd.to_datetime. This allows for proper time-series analysis and date-based sorting.

```
# 1. Calculate total spending per customer
total_spending = df.groupby('Customer ID')['Sales'].sum().reset_index()
total_spending.columns = ['Customer ID', 'total_spending']

# 2. Calculate order frequency (number of unique orders)
order_freq = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
order_freq.columns = ['Customer ID', 'order_frequency']

# 3. Calculate average order value
avg_order = df.groupby('Customer ID')['Sales'].mean().reset_index()
avg_order.columns = ['Customer ID', 'avg_order_value']

# 4. Merge all metrics into one summary table
customer_summary = total_spending.merge(order_freq, on='Customer ID').merge(avg_order, on='Customer ID')

# 5. Save the results to CSV files
df.to_csv('superstore_cleaned.csv', index=False)
customer_summary.to_csv('customer_summary.csv', index=False)

print("Files saved successfully!")
customer_summary.head(10) # This will display the first 10 rows of your new table
```

I created a new customer summary table by grouping the data by 'Customer ID' to calculate total spending, order frequency, and average order value. I successfully merged these metrics and exported the final datasets as CSV files for submission.
