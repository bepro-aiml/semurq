Task 3: Handle Missing Values  
What to do: Fill missing values in numerical columns with median and categorical columns with
mode.  
How to do it:  
• Step 1: Identify which columns have missing values (from Task 2).  
• Step 2: For numerical columns, fill with median (more robust than mean if outliers exist):

```python
# Step 1: Identify which columns have missing values (from Task 2).
#• Step 2: For numerical columns, fill with median (more robust than mean if outliers exist):
# Example for a numerical column
df['Sales'].fillna(df['Sales'].median(), inplace=True)
```

```python
# Step 3: For categorical columns, fill with mode (most frequent value):
# Example for a categorical column
df['Ship Mode'].fillna(df['Ship Mode'].mode()[0], inplace=True)
```

```python
# Step 4: Verify no missing values remain:
print(df.isnull().sum())
```

After checking with df.isnull().sum(), the dataset has 0 missing values across
all 21 columns. No filling was needed. Had there been missing values, I would
have used median for numerical columns (robust to outliers) and mode for
categorical columns (most frequent value).

Task 5: Convert Data Types  
What to do: Convert Order Date and Ship Date columns to proper datetime format.  
How to do it:

```python
# Step 1: Convert the date columns:
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
# Tool: pd.to_datetime() — converts strings to datetime objects
```

```python
#Step 2: Verify the conversion:
print(df.dtypes)
```

```python
# Step 3: Spot-check a few values to make sure dates look correct:
df[['Order Date', 'Ship Date']].head()
# Tool: pd.to_datetime() — converts strings to datetime objects
```

Task 6: Create Derived Features  
What to do: Create three customer-level summary metrics: total spending, order frequency, and
average order value.  
How to do it:

```python
# Step 1: Calculate total spending per customer:
total_spending = df.groupby('Customer ID')['Sales'].sum().reset_index()
total_spending.columns = ['Customer ID', 'total_spending']
```

```python
# Step 2: Calculate order frequency (number of orders per customer):
order_freq = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
order_freq.columns = ['Customer ID', 'order_frequency']
```

```python
# Step 3: Calculate average order value:
avg_order = df.groupby('Customer ID')['Sales'].mean().reset_index()
avg_order.columns = ['Customer ID', 'avg_order_value']
```

```python
# Step 4: Merge all into a customer summary table:
customer_summary = total_spending.merge(order_freq, on='Customer ID').merge(avg_order, on='Customer ID')
customer_summary.head(10)
```

```python
# Step 5: Save the cleaned DataFrame:
df.to_csv('superstore_cleaned.csv', index=False)
customer_summary.to_csv('customer_summary.csv', index=False)

# Tool: df.groupby(), .sum(), .nunique(), .mean(), .merge(), .to_csv()
```
