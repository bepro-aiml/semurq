```python
import pandas as pd
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
df.head(10)
```

I imported the pandas library and loaded the Superstore dataset. The head() function confirms the data has loaded correctly with columns like Ship Mode, Segment, etc.

```python
# Check shape (rows, columns)
print("Shape:", df.shape)
# Check data types and non-null counts
print("\nInfo:")
df.info()
# Summary statistics for numerical columns
print("\nDescribe:")
df.describe()
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

The dataset has [insert number] rows. I checked for missing values using isnull().sum().

```python
# Fill numerical columns with median
num_cols = df.select_dtypes(include=['number']).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Fill text columns with mode (most common value)
text_cols = df.select_dtypes(include=['object']).columns
for col in text_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Check if any missing values remain
print("Remaining Missing Values:", df.isnull().sum().sum())
```

I filled missing numerical values with the median and text values with the mode.

```python
# Count and remove duplicates
print("Duplicates found:", df.duplicated().sum())
df = df.drop_duplicates()
print("New shape:", df.shape)
```

I removed duplicate rows from the dataset.Duplicatetes found:0



```python
# Convert dates to proper format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Check data types
print(df[['Order Date', 'Ship Date']].dtypes)
```

I converted Order Date and Ship Date to datetime format.

```python
# 1. Total Spending per Customer
total_spending = df.groupby('Customer ID')['Sales'].sum().reset_index()
total_spending.columns = ['Customer ID', 'total_spending']

# 2. Order Frequency per Customer
order_freq = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
order_freq.columns = ['Customer ID', 'order_frequency']

# 3. Average Order Value per Customer
avg_order = df.groupby('Customer ID')['Sales'].mean().reset_index()
avg_order.columns = ['Customer ID', 'avg_order_value']

# Merge them together
customer_summary = total_spending.merge(order_freq, on='Customer ID')
customer_summary = customer_summary.merge(avg_order, on='Customer ID')

# Show result
customer_summary.head()

# SAVE FILES
df.to_csv('superstore_cleaned.csv', index=False)
customer_summary.to_csv('customer_summary.csv', index=False)
print("Files saved!")
```

I created customer summaries and saved the cleaned files.
