```python
import pandas as pd
import os

# The 'path' variable should have been set by a previous KaggleHub download cell
# (e.g., cell Ogv31XNZSr7y or 770c89b6).
# If you haven't run that cell yet, please do so first.

csv_file_name = "Sample - Superstore.csv"

if 'path' in locals() or 'path' in globals():
    csv_file_path = os.path.join(path, csv_file_name)
    if os.path.exists(csv_file_path):
        try:
            df = pd.read_csv(csv_file_path, encoding='latin-1')
            print(f"Loaded dataset from {csv_file_path}: {df.shape[0]} rows, {df.shape[1]} columns")
        except Exception as e:
            print(f"Error reading file '{csv_file_name}': {e}")
    else:
        print(f"Error: '{csv_file_name}' not found in path: {path}")
        print("Please ensure the file exists or check the contents of the downloaded directory.")
        if os.path.exists(path):
            print("Contents of downloaded directory:", os.listdir(path))
else:
    print("Error: 'path' variable not found. Please ensure the Kaggle dataset has been downloaded.")
    print("Run the KaggleHub download cell (e.g., cell Ogv31XNZSr7y) first to set the 'path' variable.")
```

```python
# Example for a numerical column
df['Sales'].fillna(df['Sales'].median(), inplace=True)
```

```python
# Example for a categorical column
df['Ship Mode'].fillna(df['Ship Mode'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
print(df.dtypes)
```

```python
df[['Order Date', 'Ship Date']].head()
```

```python
total_spending = df.groupby('Customer ID')['Sales'].sum().reset_index()
total_spending.columns = ['Customer ID', 'total_spending']
```

```python
order_freq = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
order_freq.columns = ['Customer ID', 'order_frequency']
```

```python
avg_order = df.groupby('Customer ID')['Sales'].mean().reset_index()
avg_order.columns = ['Customer ID', 'avg_order_value']
```

```python
customer_summary = total_spending.merge(order_freq, on='Customer ID').merge(avg_order, on='Customer ID')
customer_summary.head(10)
```

```python
df.to_csv('superstore_cleaned.csv', index=False)
customer_summary.to_csv('customer_summary.csv', index=False)
```
