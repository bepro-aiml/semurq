# From Words to Numbers: Bag of Words and TF-IDF

Machine learning models need numbers, not text. This notebook walks through two fundamental text vectorization methods:

1. **Bag of Words (BoW)** — count how many times each word appears
2. **TF-IDF** — weight words by how informative they are

We start by building BoW manually, then verify with scikit-learn.

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

## 1. Our Corpus

Five short sentences about telecom services.

```python
corpus = [
    "internet speed is very slow today",
    "my internet connection drops every hour",
    "how to change my tariff plan",
    "internet speed should be faster on this plan",
    "I want to cancel my plan and switch provider"
]

for i, doc in enumerate(corpus):
    print(f"Doc {i}: {doc}")
```

## 2. Manual Bag of Words

Steps:
1. Split each sentence into words
2. Build a vocabulary (sorted unique words)
3. For each document, count occurrences of each vocabulary word

```python
# Step 1: Tokenize
tokenized = [doc.lower().split() for doc in corpus]

# Step 2: Build vocabulary
vocab = sorted(set(word for doc in tokenized for word in doc))
print(f"Vocabulary ({len(vocab)} words): {vocab}\n")

# Step 3: Build the BoW matrix
bow_matrix = []
for doc in tokenized:
    row = [doc.count(word) for word in vocab]
    bow_matrix.append(row)

bow_df = pd.DataFrame(bow_matrix, columns=vocab, index=[f"Doc {i}" for i in range(len(corpus))])
bow_df
```

**What to notice:** Every document is now a row of numbers. Words that appear get a count >= 1; words that don't appear get 0. The matrix is *sparse* (mostly zeros).

## 3. Verify with CountVectorizer

scikit-learn's `CountVectorizer` does exactly what we did above.

```python
cv = CountVectorizer()
bow_sklearn = cv.fit_transform(corpus)

bow_sklearn_df = pd.DataFrame(
    bow_sklearn.toarray(),
    columns=cv.get_feature_names_out(),
    index=[f"Doc {i}" for i in range(len(corpus))]
)
bow_sklearn_df
```

```python
# Quick sanity check: our manual matrix should match sklearn's
# (Column order might differ, but values per document should be the same)
print("Manual vocab size:", len(vocab))
print("Sklearn vocab size:", len(cv.get_feature_names_out()))
```

## 4. TF-IDF Vectorization

**Problem with BoW:** Common words like "my", "is", "to" get high counts but carry little meaning.

**TF-IDF** fixes this by:
- **TF (Term Frequency):** How often a word appears in a document
- **IDF (Inverse Document Frequency):** Penalizes words that appear in many documents

Result: rare, specific words get **high** scores; common words get **low** scores.

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(corpus)

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray().round(3),
    columns=tfidf.get_feature_names_out(),
    index=[f"Doc {i}" for i in range(len(corpus))]
)
tfidf_df
```

## 5. High vs Low TF-IDF Scores

Let's find which words are most and least informative.

```python
# Average TF-IDF score across all documents
avg_scores = tfidf_df.mean().sort_values(ascending=False)

print("=== Highest TF-IDF (rare, informative words) ===")
print(avg_scores.head(8).to_string())
print()
print("=== Lowest TF-IDF (common, less informative words) ===")
print(avg_scores.tail(8).to_string())
```

```python
import matplotlib.pyplot as plt

# Visualize
plt.figure(figsize=(12, 5))
avg_scores.sort_values().plot(kind='barh', color='steelblue')
plt.title('Average TF-IDF Score per Word')
plt.xlabel('TF-IDF Score')
plt.tight_layout()
plt.show()
```

**Key takeaway:** Words like "internet" and "plan" appear in multiple documents, so TF-IDF gives them lower scores. Words that only appear in one document (like "cancel", "drops", "tariff") get higher scores because they are more discriminative.
