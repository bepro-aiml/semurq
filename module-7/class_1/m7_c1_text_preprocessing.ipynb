## 1. Setup

```python
import nltk
import string
import re

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

print("Setup complete.")
```

## 2. Sample Text

We will use a realistic customer complaint as our running example.

```python
text = "Uztelecom's customer service is NOT great!!! I've been waiting for 3 hours..."
print("Original text:")
print(text)
```

## 3. Step-by-Step Preprocessing

### 3.1 Lowercasing

Convert everything to lowercase so that "NOT" and "not" are treated as the same token.

```python
step1 = text.lower()
print("After lowercasing:")
print(step1)
```

### 3.2 Punctuation Removal

Strip punctuation characters. We use regex to keep only letters, digits, and spaces.

```python
step2 = re.sub(r'[^a-z0-9\s]', '', step1)
print("After punctuation removal:")
print(step2)
```

### 3.3 Tokenization

Split the string into individual tokens (words).

```python
step3 = word_tokenize(step2)
print("After tokenization:")
print(step3)
```

### 3.4 Stopword Removal

Remove common English words that usually carry little meaning ("the", "is", "for", etc.).

```python
stop_words = set(stopwords.words('english'))
print(f"Total English stopwords in NLTK: {len(stop_words)}")
print(f"Sample stopwords: {list(stop_words)[:15]}")
print()

step4 = [w for w in step3 if w not in stop_words]
print("After stopword removal:")
print(step4)
```

### The "NOT" Problem

Look carefully at the output above. The word **"not"** was removed because it is in the stopword list.

The original sentence says the service is **NOT great** -- a negative statement.
After stopword removal, we are left with words like **"great"** without the negation.

A sentiment analysis model reading the processed text would likely classify this as **positive**,
which is the opposite of what the customer meant.

**Takeaway:** Blindly removing stopwords can destroy meaning, especially for sentiment tasks.
In practice, you may want to keep negation words ("not", "no", "never", "nor") or use
models that handle raw text (like transformers).

```python
# Verify: is "not" in the NLTK stopword list?
print(f"'not' in stopwords: {'not' in stop_words}")
print(f"'no' in stopwords:  {'no' in stop_words}")
print(f"'never' in stopwords: {'never' in stop_words}")
```

### 3.5 Stemming

Reduce words to their root form. "waiting" -> "wait", "hours" -> "hour", etc.

```python
stemmer = PorterStemmer()
step5 = [stemmer.stem(w) for w in step4]
print("After stemming:")
print(step5)
```

## 4. Full Pipeline Summary

```python
print("=" * 60)
print(f"Original:             {text}")
print(f"After lowercasing:    {step1}")
print(f"After punctuation:    {step2}")
print(f"After tokenization:   {step3}")
print(f"After stopwords:      {step4}")
print(f"After stemming:       {step5}")
print("=" * 60)
```

## 5. Reusable Function

Let's wrap everything into a single function for reuse.

```python
def preprocess(text, remove_stopwords=True, stem=True, verbose=False):
    """Full text preprocessing pipeline."""
    # Lowercase
    result = text.lower()
    if verbose:
        print(f"Lowercased:  {result}")

    # Remove punctuation
    result = re.sub(r'[^a-z0-9\s]', '', result)
    if verbose:
        print(f"No punct:    {result}")

    # Tokenize
    tokens = word_tokenize(result)
    if verbose:
        print(f"Tokens:      {tokens}")

    # Stopword removal
    if remove_stopwords:
        tokens = [w for w in tokens if w not in stop_words]
        if verbose:
            print(f"No stops:    {tokens}")

    # Stemming
    if stem:
        stemmer = PorterStemmer()
        tokens = [stemmer.stem(w) for w in tokens]
        if verbose:
            print(f"Stemmed:     {tokens}")

    return tokens

# Quick test
result = preprocess(text, verbose=True)
print(f"\nFinal: {result}")
```
